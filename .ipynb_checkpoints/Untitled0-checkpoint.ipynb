```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from pandas.io.data import DataReader
from datetime import datetime
start = datetime(2010, 1, 1)
end   = datetime(2013, 1, 27)

goog = DataReader("GOOG", "google", start,end)

cierre = goog['Close'].values

len(cierre)
```

```python
ret_diario = np.zeros(len(cierre))

for i in range(1,len(cierre)):
    ret_diario[i] = ((cierre[i]/cierre[i-1])-1)*100.0
```

```python

plt.plot(ret_diario, 'k-');
```
